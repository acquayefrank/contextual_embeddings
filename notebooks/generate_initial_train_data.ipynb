```python
import json
import pandas as pd
```

```python
DATA_PATH = '../data/'
```

```python
with open(f'{DATA_PATH}_words_with_hyponyms.txt', 'r') as reader:
    words = reader.readlines()
```

```python
words = json.loads(words[0])
```

```python
max_length = 0
for index, word in enumerate(words):
    len_words = len(words[word])
    if  len_words > max_length:
        max_length = len_words
    if len_words == 100:
        print(word)
print(max_length)
```

```python
words_with_features = {}
for index, word in enumerate(words):
    temp_word_features = words[word].copy()
    if len(temp_word_features) < max_length:
        while len(temp_word_features) < max_length:
            temp_word_features += [{}]
    words_with_features[word] = temp_word_features
```

```python
df = pd.DataFrame.from_dict(words_with_features, orient='index')
```

```python
df.head()
```

```python
words_with_extracted_features = {}
for index, word in enumerate(words):
    features = [data['word'] for data in words[word]]
    if len(features) < max_length:
        while len(features) < max_length:
            features += [" "]
    words_with_extracted_features[word] = sorted(features)
```

```python
wwef_df = pd.DataFrame.from_dict(words_with_extracted_features, orient='index')
```

```python
wwef_df.head()
```

```python
all_features = []
for word in words:
    features = [data['word'] for data in words[word]]
    all_features += features
all_features = set(all_features)
```

```python
import numpy as np


fwe_df = pd.DataFrame(0, index=np.arange(len(wwef_df)), columns=sorted(list(all_features)))
```

```python
fwe_df.head()
```

```python
fwe_df.insert(loc=0, column='actual_words', value=list(wwef_df.index))
```

```python
fwe_df.head()
```

```python
fwe_df.set_index('actual_words', inplace=True)
```

```python
fwe_df.loc["aaron"]
```

```python
for index, row in fwe_df.iterrows():
    for data in wwef_df.loc[index].iteritems():
        if row.get(data[1]) is not None:
            fwe_df.at[index, data[1]] = 1
```
